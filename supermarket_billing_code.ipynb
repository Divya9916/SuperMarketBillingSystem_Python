# Supermarket Billing System

This project calculates the total bill for multiple items, applies a discount, and generates a simple receipt using Python.


```python
from datetime import datetime

print("="*50)
print("        WELCOME TO SMART BILLING SYSTEM")
print("="*50)

cart = []
subtotal = 0

while True:
    name = input("\nEnter item name: ")
    qty = int(input("Enter quantity: "))
    price = float(input("Enter price per unit: "))
    category = input("Enter category (grocery/beauty/electronics/other): ").lower()

    total = qty * price

    # Category discount
    discount = 0
    if category == "beauty":
        discount = total * 0.05
    elif category == "grocery":
        discount = total * 0.02

    final_price = total - discount
    subtotal += final_price

    cart.append([name, qty, price, total, discount, final_price])

    more = input("Do you want to add another item? (yes/no): ")
    if more.lower() != "yes":
        break


# -------- PRINT RECEIPT -------- #

print("\n\n")
print("="*60)
print("                 SMART MART")
print("            GST INVOICE / RECEIPT")
print("="*60)

now = datetime.now()
print("Date:", now.strftime("%d-%m-%Y"))
print("Time:", now.strftime("%H:%M:%S"))
print("-"*60)

print(f"{'Item':18}{'Qty':<6}{'Price':<10}{'Total':<10}{'Disc':<8}{'Final'}")
print("-"*60)

for item in cart:
    print(f"{item[0]:18}{item[1]:<6}{'₹'+format(item[2],'.2f'):<10}"
          f"{'₹'+format(item[3],'.2f'):<10}{'₹'+format(item[4],'.2f'):<8}"
          f"₹{item[5]:.2f}")

print("-"*60)

# Bill level discount
bill_discount = 0
if subtotal > 2000:
    bill_discount = subtotal * 0.10

final_amount = subtotal - bill_discount

print(f"{'Subtotal:':40} ₹{subtotal:.2f}")
print(f"{'Bill Discount:':40} -₹{bill_discount:.2f}")
print("="*60)
print(f"{'FINAL AMOUNT TO PAY:':40} ₹{final_amount:.2f}")
print("="*60)

print("        Thank you for shopping with us!")
print("              Visit Again 🙂")
print("="*60)
```
